# Formateo de JSON para la Game Jam

Enlace al dataset: [kaggle](https://www.kaggle.com/datasets/duketemon/antonyms-wordnet)

### Importar librerías

```python
import json
```

### Cargar JSON

```python
with open("antonyms.json", "r") as file:
    data = json.load(file)

print(json.dumps(data, indent = 4))
```

### Obtener claves (*keys*) sin ":"

Las claves también incluyen su **categoría gramatical** (clase de palabra/categoría morfosintáctica) y es algo que realmente no necesito (¿aunque en un futuro podría servir para dividir el juego en categorías?), así que simplemente uso la función `split()` para obtener el primer índice.  
La función `split()` devuelve una lista de elementos, de ahí a que me quede con el primer índice (la palabra en sí).

```python
for key, value in data.items():
    print(key.split(":")[0])
```

### Obtener valores (*values*)

Los valores también incluyen unos separadores extraños (no entiendo porqué hay dos): `|` y `;`. Reemplazando `|` por `;` ya solo tengo que separar en base a un solo separador.  
Como algunas palabras se quedan con valores repetidos (véase **above** que guarda **bellow** y **bellow**), lo paso a un `set` ya que no admite valores repetidos. Luego por facilidad lo vuelvo a convertir en una lista.

```python
for key, value in data.items():
    value = value.replace("|", ";")
    values = set(value.split(";"))
    values = list(values)
    print(values)
```

### Guardar en un JSON

Primero hago lo mismo que arriba pero a la vez y luego `json.dumps()` que sirve para guardar en un nuevo documento.

```python
formatted = dict()

for key, value in data.items():
    formatted.update({key.split(":")[0] : list(set(value.replace("|", ";").split(";")))})

print(formatted)
```

```python
with open("antonyms_formatted.json", "w", encoding="utf-8") as f:
    json.dump(formatted, f, ensure_ascii=False)
```
